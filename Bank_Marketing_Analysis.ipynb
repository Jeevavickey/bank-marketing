#STAGE 2

## Check Missing Values (Handling Missing Data)

```python
# Step 8: Missing value check
df.isnull().sum()
```

```python
# Step 9: duplicate
df.duplicated().sum()
```

```python
# Step 10: remove duplicates
df.drop_duplicates(inplace=True)
```

```python
# Step 11: fill the missing value numeric
num_cols=df.select_dtypes(include="number").columns

# Clean data
df[num_cols]=df[num_cols].fillna(df[num_cols].mean())
```

```python
# Step 12: unique value
df["poutcome"].unique()
```

```python
# Step 14: fill the missing value mode

df["poutcome"].fillna(df["poutcome"].mode()[0], inplace=True)

df["contact"].fillna(df["contact"].mode()[0], inplace=True)

df["month"].fillna(df["month"].mode()[0], inplace=True)
```

```python
#  Step 15: previous outcome value count
df['poutcome'].value_counts()
```

```python
# Step 16:contact value count
df["contact"].value_counts()
```

```python
# Step 17:month value count
df["month"].value_counts()
```

```python
# Step 18:Select categorical columns
cat_cols = df.select_dtypes(include='object').columns

# Clean text data
df[cat_cols] = df[cat_cols].apply(lambda x: x.str.lower().str.replace(r'[^a-zA-Z0-9\s]', '', regex=True).str.strip())
```

```python
# Step 19:null value recheck
df.isnull().sum()
```

## Creating derived columns

```python
#  Step 20: creating derived column in "age group"
df["age_group"] = pd.cut(
    df["age"].astype(int),
    bins=[18, 30, 45, 60, 80],
    labels=["Child", "Young", "Middle-aged", "Senior"]
)
df
```

##Filtering

```python
#  Step 21: filter the age above 60
filtered_df=df[df["age"].astype(int) > 60]
filtered_df
```

```python
# Step 22: groupby
df.groupby("education").agg({"balance":"mean","y":"count"})
```

#1.Univariate Analysis

```python
# Step 26: Distribution of Customer Age(Histogram Plot)
plt.figure(figsize=(10,6))
sns.histplot(df["age"],bins=20,color="LightCoral",edgecolor="IndianRed",kde=True)
plt.title("Age Distribution",color="red",fontstyle="italic")
plt.xlabel("Age",color="blue")
plt.ylabel("Frequency",color="blue")
plt.show()
```

### **Interpretation**

The age distribution shows that most customers belong to the middle-age group.
Very young and elderly customers are fewer, indicating a concentrated working-age population.

```python
# Step 27:Distribution of Job Categories(bar chart)
plt.figure(figsize=(8,5))
sns.countplot(df["job"],color="PaleTurquoise",edgecolor="Aqua")
plt.title("job distribution",color="green",fontstyle="italic")
plt.xlabel("Job",color="MediumVioletRed")
plt.ylabel("Frequency",color="MediumVioletRed")
plt.show()
```

### **Interpretation**

The bar chart shows the frequency of customers across different job categories.

A few job roles have higher representation, while some jobs appear less frequently, indicating an uneven job distribution in the dataset

```python
#Step 28: Distribution of Account Balance(box plot)
plt.figure(figsize=(8,5))
sns.boxplot(x=df["balance"],color="DarkSlateGray")
plt.title("Balance Distribution",color="red",fontstyle="italic")
plt.show()
```

### **Interpretation**

The box plot shows the median balance, spread, and outliers.

The presence of outliers indicates that some customers have very high or very low balances compared to the majority.

```python
# Step 29:Distribution of Subscription Status(pie chat)
plt.figure(figsize=(5,4))
df["y"].value_counts().plot.pie(autopct=("%1.1f%%"),colors=["Teal","DarkOrange"])
plt.title("Subscription Status",color="Indigo")
plt.ylabel("")
plt.show()
```

### **Interpretation**

The pie chart shows the proportion of customers who subscribed (Yes) and did not subscribe (No).

One category occupies a larger share, indicating that most customers belong to that subscription status

#2. Bivariate Analysis

```python
# Step 30:Age and Account Balance Relationship(scater plot)
plt.figure(figsize=(8,5))
sns.scatterplot(data=df,x="age",y="balance",color="DarkGoldenrod")
plt.title("Age vs Balance",color="SlateBlue")
plt.show()
```

### **Interpretation**

The scatter plot shows the relationship between age and account balance.

There is no strong linear pattern, but balances tend to vary widely across different age groups.

```python
# Step 31:Relationship Between Job and Subscription Outcome(bar chart)
plt.figure(figsize=(8,5))
sns.countplot(data=df,x="job",hue="y",palette=["Goldenrod","Maroon"])
plt.title("Job vs Subscription",color="RebeccaPurple")
plt.xticks(rotation=70,color="SaddleBrown")
plt.show()
```

### **Interpretation**

The chart compares subscription status (Yes/No) across different job categories.

It helps identify which job groups have higher or lower subscription rates, showing how subscription behavior varies by occupation

```python
# Step 32:Call Duration by Subscription Status(box plot)
plt.figure(figsize=(8,5))
sns.boxplot(x="y",y="duration",data=df,palette=["LightSlateGray","DarkGray"])
plt.title("Subscription vs Call Duration",color="Maroon")
plt.show()
```

### **Interpretation**

The box plot compares call duration for customers who subscribed (Yes) and did not subscribe (No).

Customers who subscribed generally have longer call durations, indicating that longer interactions are associated with successful subscriptions.

#3.Multivariate Analysis

```python
# Step 33:Correlation Matrix of Numerical Variables(matrix)
plt.figure(figsize=(8,5))
sns.heatmap(df.corr(numeric_only=True),annot=True,cmap="coolwarm")
plt.title("Correlation Heatmap",color="DarkSlateGray")
plt.tight_layout()
plt.show()
```

### **Interpretation**

The heatmap shows the strength and direction of relationships between numerical variables.

Variables with strong positive or negative correlations are highlighted, helping identify features that are closely related

```python
#Step 34:Pairwise Analysis of Customer Age,Balance and Call Duration
plt.figure(figsize=(8,5))
sns.pairplot(df[["age","balance","duration","y"]],hue="y",palette={"yes":"DarkSlateGray","no":"Peru"})
plt.suptitle("Pair Plot of Age, Balance, and Duration")
plt.show()
```

### **Interpretation**

The pair plot shows pairwise relationships between age, balance, and duration, separated by subscription status (y).

Customers who subscribed (yes) and did not subscribe (no) are colored differently (Salmon vs Crimson).

It helps visualize patterns, correlations, and distribution differences between the two groups across all numerical variables


```python
# Step 35: Monthly Distribution of Previous Campaign Outcomes(Stacked bar chart)
plt.figure(figsize=(10, 7))
pd.crosstab(df["month"],df["poutcome"]).plot(kind="bar",stacked=True,color=["teal","DeepPink","Gold","LightCoral"])
plt.title("Month vs Previous Outcome (Stacked)",color="MidnightBlue")
plt.xlabel("Month",color="Indigo")
plt.ylabel("Count",color="Indigo")
plt.legend(title="Previous Outcome")
plt.show()
```

### **Interpretation**


The chart shows how previous campaign outcomes are distributed across different months.

It helps identify month-wise variations in campaign success, failure, or unknown outcomes

```python
# Step36:Distribution of Account Balance Across Marital Groups(violinplot)
plt.figure(figsize=(8, 5))
sns.violinplot(x="marital",y="balance",data=df,palette="husl")
plt.title("Marital Status vs Balance", color="SeaGreen")
plt.xlabel("Marital Status",color="DarkSlateGray")
plt.ylabel("Balance",color="DarkSlateGray")
plt.show()
```

### **Interpretation**

The violin plot shows the distribution of account balance across different marital statuses.

It highlights differences in balance spread and density, indicating how balance varies among marital groups.
